```python
# Dependencies
import pandas as pd
import os
import csv
import numpy as np
```

```python
# load CSV
csvpath=os.path.join("Resources","2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas
clean_coders_pd = pd.read_csv(csvpath,dtype ='object')
clean_coders_pd.head()
```

```python
# Inspect all columns
clean_coders_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
best_clean_coders_pd = clean_coders_pd.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
best_clean_coders_pd.head()
```

```python
# Change "0" to "No" and "1" to "Yclean_coders_pd = clean_coders_pd.replace({0: "No", 1: "Yes"})
# Hint use the df.replace function
best_clean_coders_pd = best_clean_coders_pd.replace({"0": "No", "1": "Yes"})
best_clean_coders_pd.head()
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
TotalResponses =len(best_clean_coders_pd)
TotalResponses
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
Best_Attended = best_clean_coders_pd.loc[best_clean_coders_pd["AttendedBootcamp"] == "Yes", :]
Best_Attended
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it

Best_Attended[("Age")] = pd.to_numeric(Best_Attended["Age"])
AvgAge = Best_Attended["Age"].mean()
AvgAge
```

```python
# Calculate how many people attended a bootcamp
Attendees = Best_Attended["AttendedBootcamp"].count()
Attendees
```

```python
# Count the values of the AttendedBootcamp column
NumAttended = Best_Attended["SchoolDegree"].value_counts()
NumAttended
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

SchoolDegree_df =  Best_Attended.loc[(Best_Attended["SchoolDegree"] == "bachelor's degree") |
                                    (Best_Attended["SchoolDegree"] == "master's degree (non-professional)") |
                                     (Best_Attended["SchoolDegree"] == "professional degree (MBA, MD, JD, etc.)") |
                                     (Best_Attended["SchoolDegree"] == "associate's degree") |
                                     (Best_Attended["SchoolDegree"] == "Ph.D.") , :]
SchoolDegree_df["SchoolDegree"].value_counts()
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees

NumDegrees = SchoolDegree_df["SchoolDegree"].count()
NumDegrees
```

```python
# Calculate percentage of people who attended with degrees
PercDegrees = (NumDegrees/ Attendees)*100
PercDegrees
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
Gender = Best_Attended["Gender"].value_counts()
Gender
```

```python
# Calculate percentage of respondents who attended a bootcamp
PercAttend =(Attendees/TotalResponses)*100
PercAttend
```

```python
#df['Employer'] = df['Employer'].replace(
    #{'Self Employed': 'Self-Employed', 'Self': 'Self-Employed'})

Best_Attended["Gender"] = Best_Attended["Gender"].replace ({"genderqueer": "non-binary", "trans":"non-binary", "agender":"non-binary"})
Best_Attended["Gender"].value_counts()
```

```python
# Calculate percentage of respondents belonging to each gender
Male_df = Best_Attended.loc[Best_Attended["Gender"]=="male"]
NumMale = Male_df["Gender"].count()
PercMale = (NumMale/Attendees)*100

Female_df = Best_Attended.loc[Best_Attended["Gender"]=="female"]
NumFemale = Female_df["Gender"].count()
PercFemale = (NumFemale/Attendees)*100


NonBinary_df = Best_Attended.loc[Best_Attended["Gender"]=="non-binary"]
NumNonB = NonBinary_df["Gender"].count()
PercNonB = (NumNonB/Attendees)*100
```

```python
# Calculate percentage of attendees with a college degree

PercDegrees
```

```python
# Calculate average post-bootcamp salary
## Drop all rows with missing information df = df.dropna(how='any')
Best_Attended[("BootcampPostSalary")] = pd.to_numeric(Best_Attended["BootcampPostSalary"])
AvgPostBoot = Best_Attended["BootcampPostSalary"].mean()
AvgPostBoot
```

```python
# Create a new table consolidating above calculations
Finaltable_df = pd.DataFrame({"Total Surveyed":[TotalResponses],
                 "Total Bootcamp attendees":[Attendees],
                 "Percent attended bootcamp":[PercAttend],
                 "Avg. Age":[AvgAge],
                 "Has a degree":[PercDegrees],
                 "% Male":[PercMale],
                 "% Female":[PercFemale],
                 "% Non Gender Specific":[PercNonB],
                 "Average Post Bootcamp Salary":[AvgPostBoot]})
Finaltable_df
```
