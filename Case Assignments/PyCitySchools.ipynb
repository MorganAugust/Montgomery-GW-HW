# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import os
import csv
import numpy as np
```

```python
# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset (consider using a left join)
SchoolandStudent_df = pd.merge(school_data, student_data, on="school_name", how="left")
SchoolandStudent_df
```

## District Summary

```python
# Calculate the Totals (Schools and Students)
school_list = SchoolandStudent_df["school_name"].unique()
Total_Schools = len(school_list)
Total_Schools
```

```python
num_students = SchoolandStudent_df["Student ID"].unique()
Total_Students = len(num_students)
Total_Students
```

```python
# Calculate the Total Budget
Total_Budget = school_data["budget"].sum()
Total_Budget
```

```python
# Calculate the Average Scores
Average_Math = student_data["math_score"].mean()
Average_Reading = student_data["reading_score"].mean()
print(Average_Math)
print(Average_Reading)
```

```python
# Calculate the Percentage Pass Rates
```

```python
# Calculate the Percentage Pass Rates - Math 
PassingMath = SchoolandStudent_df.loc[SchoolandStudent_df["math_score"]>69]
len(PassingMath)
```

```python
# Calculate the Percentage Pass Rates - Math
PercentPassingMath =((len(PassingMath)/Total_Students)*100)
PercentPassingMath
```

```python
# Calculate the Percentage Pass Rates - Reading
PassingReading = SchoolandStudent_df.loc[SchoolandStudent_df["reading_score"]>69]
#len(PassingReading)
PassingReading.head()
```

```python
# Calculate the Percentage Pass Rates - Reading
PercentPassingReading = ((len(PassingReading)/Total_Students)*100)
PercentPassingReading
```

```python
# Calculate the Percentage Pass Rates - Overall
PassedBoth = SchoolandStudent_df.loc[PassingReading.index & PassingMath.index]
```

```python
PercentPassingOverall = ((len(PassedBoth)/Total_Students)*100)
PercentPassingOverall
```

```python
# Minor Data Cleanup
# Display the data frame
District_Summary = pd.DataFrame({"Total Schools": [Total_Schools],
                     "Total Students": [Total_Students],
                     "Total Budget": [Total_Budget],
                     "Average Math Score": [Average_Math],
                     "Average Reading Score": [Average_Reading],
                     "% Passing Math": [PercentPassingMath],
                     "% Passing Reading": [PercentPassingReading],
                     "% Overall Passing Rate": [PercentPassingOverall],
                    })

District_Summary["Total Budget"] =District_Summary["Total Budget"].map("${0:,.2f}".format)
District_Summary["Total Students"] =District_Summary["Total Students"].map("{0:,.0f}".format)
#District_Summary["% Passing Math"] = District_Summary["% Passing Math"].map("{0:.2f}%".format)


District_Summary
```

## School Summary

```python
# Determine the School Type
BySchool = school_data.set_index("school_name")["type"]
BySchool.head()
```

```python
# Calculate the total student count
SchoolGroup = SchoolandStudent_df.groupby(["school_name"])
TotalStudentsbySchool = SchoolGroup["Student ID"].count()
TotalStudentsbySchool
```

```python
SchoolandStudent_df["student_name"].count()
```

```python
# Calculate the total school budget and per capita spending
# per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget = school_data.groupby(["school_name"]).mean()["budget"]
per_school_budget 
```

```python
# Calculate the per capita spending
per_capita = per_school_budget/TotalStudentsbySchool
per_capita
```

```python
# Calculate the average test scores
AvgReadingbySchool = SchoolGroup["reading_score"].mean()
AvgReadingbySchool

AvgMathbySchool = SchoolGroup["math_score"].mean()
AvgMathbySchool
```

```python
# Calculate the passing scores by creating a filtered data frame
#%PassingMath
#%PassingReading
#%OverallPassing

PassingReadingbySchool = PassingReading.groupby(["school_name"]).count()["reading_score"]
PercentPassingReadingbySchool = (PassingReadingbySchool/TotalStudentsbySchool)*100

PassingMathbySchool = PassingMath.groupby(["school_name"]).count()["math_score"]
PercentPassingMathbySchool = (PassingMathbySchool/TotalStudentsbySchool)*100

OverallPassingbySchool = PassedBoth.groupby(["school_name"]).count()["reading_score"]
PercentPassingOverallbySchool = (OverallPassingbySchool/TotalStudentsbySchool)*100
```

```python
# Convert to data frame
# Minor data munging
School_Summary = pd.DataFrame({"School Type": BySchool,
                     "Total Students": TotalStudentsbySchool,
                     "Total Budget": per_school_budget,
                    "Per Student Budget": per_capita,
                     "Average Math Score": AvgMathbySchool,
                     "Average Reading Score": AvgReadingbySchool,
                    "% Passing Math":PercentPassingMathbySchool,
                    "% Passing Reading": PercentPassingReadingbySchool,
                    "% Overall Passing Rate": PercentPassingOverallbySchool,
                    })

School_Summary["Total Budget"] =School_Summary["Total Budget"].map("${0:,.2f}".format)
School_Summary["Per Student Budget"] =School_Summary["Per Student Budget"].map("${0:,.2f}".format)
#School_Summary["Total Students"] =School_Summary["Total Students"].map("{0:,.0f}".format)
#Display the data frame
School_Summary

```
